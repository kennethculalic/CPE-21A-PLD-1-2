Example 1

Write a program that determines if the input age is qualified to vote or not. The qualifying age is 18 years old and above. 

```python
age = int(input("age: "))
if age >= 18:
    print("You are qualified!")
else:
    print("You are a minor")
```

Example 2

Write a program that determines if the input number is POSITOVE or NEGATIVE. Consider 0 as positive (considering that it contains no negative sign.)

```python
num = int(input("number: "))
if num >= 0:
    print("Positive")
else:
    print("Negative")
```
